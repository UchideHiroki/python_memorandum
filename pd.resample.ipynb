pd.resampleはpd.Timestamp型のindexや列を集約したり広げたりする。  

デフォルトの設定  
`DataFrame.resample(rule, how=None, axis=0, fill_method=None, closed=None, label=None, convention='start', kind=None, loffset=None, limit=None, base=0, on=None, level=None)[source]`

主要な引数一覧 

引数 | 型 | 内容 |
:---: | :---: | :---
rule | string | resampleする方法、指定出来る文字一覧は後述  
closed | {'right', 'left'} | 'left': 最初の行から見て下にresample, 'right': 最初の行から見て上にresample
label | {'right', 'left'} | 'left': 集約した行の始点が表示index, 'right': 集約した行の終点が表示index 一部ruleではrightしか指定出来ない(後述)
on | string | DFを特定の列でresampleする場合において、指定する列名
level | int | DFをmultiindexでresampleする場合において、何番目のindexでresampleするか

よく使いそうなresampleする方法一覧

表記 | 区切り単位
:---: | :---
B | 営業日
D | 日付
W | 週
M | 月
SM | 15日、月末
SMS | 1日、15日
Q | 四半期
Y | 年
H | 時間
T | 分
S | 秒

参考URL: [Time Series / Date functionality](http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases)

後続する関数

| 関数 | 説明 |
|:--|:--|
| first | 未来方向に最も近い値 |
| last | 過去方向に最も近い値 |
| bfill | backward fill, 未来方向に最も近い値 （`NaN`は埋める） |
| ffill | forward fill, 過去方向に最も近い値 （`NaN`は埋める） |
| count | 値の個数 |
| unique | ユニークな値の個数 |
| max | 最大値 |
| min | 最小値 |
| mean | 平均値 |
| median | 中央値 |
| sum | 合計値 |
| var | 分散 |
| std | 標準偏差 |
| ohlc | 始値(opning)、最高値(highest)、最安値(lowest)、終値(closing) |
| pad | = ffill |

resampleした後はrollingで移動平均取りたくなったりするが、詳しく調べるのはまた今度

```python
import pandas as pd
from numpy.random import *
index = pd.date_range('1/1/2000', periods=100, freq='D')
```

```python
%matplotlib inline
df = pd.DataFrame({'value': [i + j for i, j in zip(range(100), randn(100))]}, index=index)
```

```python
df.plot()
```

```python
df['7day_walk_mean'] = df['value'].rolling(window=7).mean()
```

```python
df.plot()
```
